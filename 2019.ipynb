```python
#Import files for zurich

b_trips = pd.read_csv("b_trips.txt")
b_routes = pd.read_csv("b_routes.txt", usecols=np.r_[0:5])
b_stops = pd.read_csv("b_stops.txt", usecols=np.r_[0:6])
b_transfers = pd.read_csv("b_transfers.txt")
b_stop_times = pd.read_csv("b_stop_times.txt", usecols=np.r_[0:6])
b_calendar = pd.read_csv("b_calendar.txt")#, usecols=np.r_[0:1, 8:10])
b_shapes = pd.read_csv("b_shapes.txt")

```
